## 3. National test (*)
Read in the file [NPvt19Ma2A.txt][files] and [NPvt19Ma2C.txt][files] in Python. Use **matplotlib** to plot pie charts for each grade categories in each file. (*)

[files]: https://github.com/kokchun/Programmering-med-Python-21/tree/main/Files

<details>

<summary>Hint </summary>

Use **subplot** in **matplotlib**

</details>
<br>
<details>

<summary>Answer </summary>

<img align="left" src="../assets/NP.png" width="300"/>

</details>

````python
import matplotlib.pyplot as plt

path_1 = "../Files/NPvt19Ma2A.txt"
path_2 = "../Files/NPvt19Ma2C.txt"

# import the first text data
with open (path_1, "r") as f1:
    # read each line to a list of lines
    y1 = [line.strip("\n") for line in f1.readlines()]
    # contruct a list to contain all the percentage numbers.
    y1_number = []

    # loop through each line in y1 list
    for label_number in y1:
        # get the second elment
        number = label_number.split()[1]
        # remove the %
        number = number.strip("%")
        # put each number into the numbers list, which is going to be y in plot.
        y1_number.append(number) 
    print(y1_number)
    
mylabels = ["A", "B", "C", "D", "E", "F"]
myexplode = [0, 0, 0, 0, 0, 0.1]
plt.subplots(1,2, figsize =(10, 7))

#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1) 


# labels is a list of sequence of strings which sets the label of each wedge.
# color attribute is used to provide color to the wedges.
# autopct is a string used to label the wedge with their numerical value.
# shadow is used to create shadow of wedge.
plt.pie(y1_number, labels = mylabels, explode = myexplode, shadow = True, autopct='%1.1f%%')
plt.legend()  # with explanations of all the labels.

# import the second text data
with open(path_2, "r") as f2:
    y2 = [line.strip("\n") for line in f2.readlines()]
    y2_number = []

    for label_number in y2:
        number = label_number.split()[1]
        number = number.strip("%")
        y2_number.append(number) 
    print(y2_number)
 
mylabels = ["A", "B", "C", "D", "E", "F"]
myexplode = [0, 0, 0, 0, 0, 0.1]


plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.

plt.pie(y2_number, labels = mylabels, explode = myexplode, shadow = True, autopct='%1.1f%%')
plt.legend()


plt.show() 
````

## 4. Dice roll experiment (**)
Simulate 10, 100, 1000, 10000, 100000 dice rolls and count the freqencies and probabilities for each number in each simulation. Create a new text file using Python with the name "simulation.txt" and write the results to that text file.(**)

````python
import random as rnd
numbers_of_dices = [10, 100, 1000, 10000, 100000]
dices = list(range(1,7))

with open("../Files/simulation.txt", "w") as f_dice:
    
    for n_of_times in numbers_of_dices:
        output = [rnd.randint(1,6) for _ in range(n_of_times)]
        #it's just a variable like any other, but by convention it means that you don't intend to use that value, just read it and ignore it.
        for dice in dices:
            f_dice.write(f"In total {n_of_times} dices: the frequency of {dice} is: {output.count(dice)}. And the probability is: {(output.count(dice)/n_of_times)*100:0.2f}%\n\n")


````
