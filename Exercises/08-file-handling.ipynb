````python
# In this list, only the last row don't have "\n" in the ending. 
# After sorting, this last row will go to the middle of the list. 
# Lacking a ending \n makes it tightly attached to the next row.
# One way is to add \n to that row in the end.
# In the next block, I will show by remove all \n.

path = "../Files/test_result.txt"
path_new = "../Files/test_result_new.txt"

with open(path, "r", encoding = "utf8") as f1, open(path_new, "w",  encoding = "utf8") as f2:
    text = f1.readlines()
    print(text)
    text.sort()
    print(text)
    
    f2.write("Sorted alphabetically: \n\n")
    for name_score in text:
        if name_score != "Fredrika Ulven 10":
            f2.write(f"{name_score}")
        if name_score == "Fredrika Ulven 10":
            name_score = name_score + "\n"
            f2.write(f"{name_score}")
            
````
