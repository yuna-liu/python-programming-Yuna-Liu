```python
import numpy as np
from math import sqrt 
from math import pi
import matplotlib.pyplot as plt

class NormalDist():
    def __init__ (self, mu:float, sigma:float) -> None:
        self.mu = mu
        self.sigma = sigma

    def pdf(self,x:float) -> float:
        prob_density = (1/(self.sigma*sqrt(2*pi))) * np.exp(-0.5*((x-self.mu)/self.sigma)**2)
        return prob_density

    def plot_pdf(self, x:float) -> None:
        count, bins, ignored = plt.hist(x,30,density=True)
        plt.plot(bins,self.pdf(bins), linewidth=2, color="r")
        plt.show()

    def __add__(self, other: "NormalDist") -> "NormalDist":
        mu_add = self.mu + other.mu
        sigma_add = self.sigma + other.sigma
        return NormalDist(mu_add, sigma_add)

    def __repr__(self):
        print(f"Normal distribution (mean, standard deviation): N({self.mu},{self.sigma})")

norm = NormalDist(0,1) # standard normal dist

print(f"The probability density at x= 1 is: {norm.pdf(1)}") # så får du f(1)

x = np.random.normal(0, 1, 1000)
print(x)
norm.plot_pdf(x) # ger en plot
```
