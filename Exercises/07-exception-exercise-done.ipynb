## 1. Find errors (*)

Find the errors in this code to compute the distance between the point $(x,y)$ and the origin in a cartesian coordinate system.

```python
impor numpy as np

def distance(x,y)
    reurn np.sqrt(x+y)

print(distance([0.5, 0.5]))

```

<details>

<summary>Hint </summary>

Also look for logical errors. 

</details>
<br>
<details>

<summary>Answer </summary>

```
The distance between (0.5, 0.5) and origin is around 0.707.
```

</details>


````python
import numpy as np

def distance(x,y):
    return np.sqrt(x**2+y**2)

print(f"The distance between (0.5, 0.5) and origin is around {distance(0.5,0.5):.3f}.")
````

## 2. Find errors (*)

Find the errors in this code. Just change the function, don't touch the test program.

```python

def is_fourdigit(number):
    if number//1000 < 10
        return true
    else 
        return false

# test program
test_numbers = [231, 3124, -4124, -1000,-999, 1001, 10000, -10000, 999]

for number in test_numbers:
    if is_fourdigit(number):
        print(f"{number} is four-digit")
    else:
        print(f"{number} is not four-digit")

```

<details>

<summary>Hint </summary>

Print out and test different cases to build up the condition in the if-statement

</details>
<br>
<details>

<summary>Answer </summary>

```

231 is not four-digit
3124 is four-digit
-4124 is four-digit
-1000 is four-digit
-999 is not four-digit
1001 is four-digit
10000 is not four-digit
-10000 is not four-digit
999 is not four-digit

```

</details>


````python
def is_fourdigit(number):
    if  0 < abs(number//1000) < 10:    # // the reminder after division... % mod 取整
        return True
    else: 
        return False

# test program
test_numbers = [231, 3124, -4124, -1000,-999, 1001, 10000, -10000, 999]

for number in test_numbers:
    if is_fourdigit(number):
        print(f"{number} is four-digit")
    else:
        print(f"{number} is not four-digit")
````

## 3. Tram (*)

Kokchun is a **clumpsy** computer user that doesn't take trams too often. Write a program to prompt the user for: 
- number of times he/she wants to take tram in one month
- cost for one ticket
- cost for monthly card

The program should calculate if it's worth for him to buy  monthly card or not. Make the program user friendly with clear error messages and ask again in case of input errors. 

<details>

<summary>Hint </summary>

Raise user friendly exceptions to guide the user to give right input format.

</details>
<br>
<details>

<summary>Answer </summary>

For example: 

```
How many times do you take tram in one month? -23
Number of times you take tram must be between 0 and 100

How much does one ticket cost? (kr) 1000
One ticket must cost between 0 and 100 kr

How many times do you take tram in one month? 23
How much does one ticket cost? (kr) 35
How much does one month card cost? (kr) 600

Cost with one-time tickets 805.0
Cost with monthly card 600.0
It's worth to buy a monthly card

```

</details>


````python
while True:
    n_of_times = input ("How many times do you take tram in one month?")
    
    try:
        n_of_times = float(n_of_times)
        
        if not 0 <= n_of_times <= 100:
            raise ValueError(f"Number of times you take tram must between 0 and 100, not {n_of_times}")
        break
        
    except ValueError as err:
        print(err)

while True:
    cost_one_ticket = input("How much does one ticket cost? (kr)")
    
    try:
        
        cost_one_ticket = float(cost_one_ticket)
        
        if not 0 <= cost_one_ticket <= 102:
            raise ValueError(f"One ticket must cost between 0 and 102 kr")
        break
        
    except ValueError as err:
        print(err)


while True:
    
    cost_monthly_card = input ("How much does one month card cost? (kr)")

    try:
        cost_monthly_card = float(cost_monthly_card)
        
        if not 0 <= cost_monthly_card <= 1825:
            raise ValueError(f"One ticket must cost between 0 and 1825 kr")
        break
    
    except ValueError as err:
        print(err)

total_fee_month_single_ticket = n_of_times*cost_one_ticket
if total_fee_month_single_ticket > cost_monthly_card:
    print("Hi, I would like to buy monthly ticket card")
elif total_fee_month_single_ticket < cost_monthly_card:
    print("Hi, I would like to buy single ticket every trip")
else:
    print("It does not matter which to choose.")
````
